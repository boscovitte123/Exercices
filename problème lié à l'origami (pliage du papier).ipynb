【Problème 1】Implémentation utilisant l'opérateur arithmétique pour les puissances

```python
# Code pour calculer l'épaisseur lorsque le papier est plié une fois

THICKNESS = 0.00008 # Déclaration de la constante (Epaisseur initiale)

folded_thickness = THICKNESS * 2**43

print(f"Épaisseur : {folded_thickness} mètres ")
```

[Problème 2] Conversion d'unité

```python
# Convertir des mètres en kilomètres avec 2 décimales

folded_thickness_in = folded_thickness/1000 # Convertion
print(f"Épaisseur : {folded_thickness_in:.2f} kilomètres")
```

[Problème 3] Créé à l'aide d'une instruction for

```python
nbr_de_plis = [0] * 42

THICKNESS = 0.00008
# Constante
folded_thickness = THICKNESS * 2  # Premier pli

for fold in nbr_de_plis:
   folded_thickness = folded_thickness * 2
folded_thickness_in = folded_thickness/1000
print(f"Épaisseur : {folded_thickness_in:.2f} kilomètres")
```

[Problème 4] Comparaison des temps de calcul

Première Méthode

```python
import time
start = time.time()

THICKNESS = 0.00008 # Déclaration de la constante (Epaisseur initiale)

folded_thickness = THICKNESS * 2**43

folded_thickness_in = folded_thickness/1000 # Convertion
print(f"Épaisseur : {folded_thickness_in:.2f} kilomètres")
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

Deuxieme Méthode

```python
import time
start = time.time()
nbr_de_plis = [0] * 42

THICKNESS = 0.00008
# Constante
folded_thickness = THICKNESS * 2  # Premier pli

for fold in nbr_de_plis:
   folded_thickness = folded_thickness * 2
folded_thickness_in = folded_thickness/1000
print(f"Épaisseur : {folded_thickness_in:.2f} kilomètres")
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

**Remarque:**
Le tempts d'exécution de la deuxième méthode est supérieur à celui de la première , de là on peut dire que la première méthode s'excécute plus vite que la deuxième

Augmentons le nombres de plies à 100 fois

```python
import time
start = time.time()

THICKNESS = 0.00008 # Déclaration de la constante (Epaisseur initiale)

folded_thickness = THICKNESS * 2**100

folded_thickness_in = folded_thickness/1000 # Convertion
print(f"Épaisseur : {folded_thickness_in:.2f} kilomètres")
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

[Problème 5] Enregistrer dans la liste

```python
nbr_de_plis = [0.00008]  # Épaisseur initiale du papier (premier pli)
n = 44  # Nombre total de plis

for fold in range(1, n):
    folded_thickness = nbr_de_plis[-1] * 2  # Calcul de l'épaisseur après le pliage
    nbr_de_plis.append(folded_thickness)  # Ajout de la nouvelle épaisseur à la liste

print(f"Nombre de plis : {len(nbr_de_plis)}")
print(f"Épaisseur finale : {nbr_de_plis[-1]:.2f} mètres")
```

[Problème 6] Affichage du graphique linéaire

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Afficher les graphiques. Avec le titre et les noms d'étiquette d'axe.
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(nbr_de_plis)  # Mettez le nom de la variable de la liste dans 'Nom de la liste'.
plt.show()
```

L'axe horizontal du graphique représente le "nombre de plis", ce qui signifie le nombre de fois que le papier a été plié.

L'axe vertical du graphique représente "l'épaisseur" du papier en mètres.

En examinant le graphique, vous pouvez voir que l'épaisseur du papier augmente de manière exponentielle à chaque pli. Au début, l'épaisseur est très faible, mais à mesure que vous effectuez plus de plis, elle augmente de manière significative. Cela correspond au comportement attendu, car chaque pli double l'épaisseur du papier par rapport à l'étape précédente.
